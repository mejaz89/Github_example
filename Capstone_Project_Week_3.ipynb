```python
df['Latitude'] = 1.000000
df['Longitude'] = 2.000000    
df
```

```python
range(df.shape[0])
    
```

```python
for n in range(df.shape[0]):
    for m in range(df1.shape[0]):
        if df['PostalCode'][n] == df1['Postal Code'][m]:
            df['Latitude'][n] = df1['Latitude'][m]
            df['Longitude'][n] = df1['Longitude'][m]
```

# Part-3

```python
import folium
```

```python
toronto_longitude = -79.347015
toronto_latitude = 43.651070
```

```python
map_toronto = folium.Map(location=[toronto_latitude, toronto_longitude], zoom_start=10)

for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  


map_toronto
```
